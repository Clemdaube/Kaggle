```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
train_df = pd.read_csv("train.csv", header=0)
train_df.head()
```

```python
test_df = pd.read_csv("test.csv", header=0)
test_df.head()
```

```python
#Convert 'Sex' into a numeric feature
data = [train_df, test_df]
genders = {"male": 0, "female": 1}

for dataset in data:
    dataset['Sex'] = dataset['Sex'].map(genders)

#Fill the two missing values of 'Embarked' with the most common one and convert it to a numeric feature
for dataset in data:
    dataset['Embarked'] = dataset['Embarked'].fillna("S")
    
ports = {"S": 0, "C": 1, "Q": 2}

for dataset in data:
    dataset['Embarked'] = dataset['Embarked'].map(ports)
    

#Adding a column named 'relatives' to the train and the test datasets
for dataset in data:
    dataset['relatives'] = dataset['SibSp'] + dataset['Parch']
    dataset.loc[dataset['relatives'] > 0, 'not_alone'] = 0
    dataset.loc[dataset['relatives'] == 0, 'not_alone'] = 1
    dataset['not_alone'] = dataset['not_alone'].astype(int)

#Extract the titles of the names and transform them into numeric values, drop the names
titles = {"Mr": 1, "Miss": 2, "Mrs": 3, "Master": 4, "Rare": 5}

for dataset in data:
    # extract titles
    dataset['Title'] = dataset["Name"].str.extract(' ([A-Za-z]+)\.', expand=False)
    # replace titles with a more common title or as Rare
    dataset['Title'] = dataset['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr',\
                                            'Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataset['Title'] = dataset['Title'].replace('Mlle', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Ms', 'Miss')
    dataset['Title'] = dataset['Title'].replace('Mme', 'Mrs')
    # convert titles into numbers
    dataset['Title'] = dataset['Title'].map(titles)
    # filling NaN with 0, to get safe
    dataset['Title'] = dataset['Title'].fillna(0)
train_df = train_df.drop(['Name'], axis=1)
test_df = test_df.drop(['Name'], axis=1)

#Extract the deck letter and transform it into numeric values, drop 'Cabin'
deck = {"A": 1, "B": 2, "C": 3, "D": 4, "E": 5, "F": 6, "G": 7, "U": 8}
data = [train_df, test_df]

for dataset in data:
    dataset['Cabin'] = dataset['Cabin'].fillna("U0")
    dataset['Deck'] = dataset['Cabin'].map(lambda x: re.compile("([a-zA-Z]+)").search(x).group())
    dataset['Deck'] = dataset['Deck'].map(deck)
    dataset['Deck'] = dataset['Deck'].fillna(0)
    dataset['Deck'] = dataset['Deck'].astype(int)# we can now drop 'Cabin'
train_df = train_df.drop(['Cabin'], axis=1)
test_df = test_df.drop(['Cabin'], axis=1)

#Transform 'Fare' in integers
train_df['Fare'] = train_df['Fare'].fillna(0)
train_df['Fare'] = train_df['Fare'].astype(int)

test_df['Fare'] = test_df['Fare'].fillna(0)
test_df['Fare'] = test_df['Fare'].astype(int)


#Since it contain too many distinct variable, 'Ticket' will be dropped from the dataset
train_df = train_df.drop(['Ticket'], axis=1)
test_df = test_df.drop(['Ticket'], axis=1)

#Since it's useless, drop 'PassengerID' from the train dataset
train_df = train_df.drop(['PassengerId'], axis=1)

#'Age' distribution is a bit skewed so we could use the median to fill out missing values
#We're gonna use a random sampling distribution to be more accurate
age_values = train_df['Age'].dropna().values    

train_df['Age'] = train_df['Age'].apply(lambda x: np.random.choice(age_values) if pd.isnull(x) else x)

age_values_test = test_df['Age'].dropna().values    

test_df['Age'] = test_df['Age'].apply(lambda x: np.random.choice(age_values_test) if pd.isnull(x) else x)
```

```python
#Create age and fare categories within the two datasets
data = [train_df, test_df]
for dataset in data:
    dataset['Age'] = dataset['Age'].astype(int)
    dataset.loc[ dataset['Age'] <= 11, 'Age'] = 0
    dataset.loc[(dataset['Age'] > 11) & (dataset['Age'] <= 18), 'Age'] = 1
    dataset.loc[(dataset['Age'] > 18) & (dataset['Age'] <= 22), 'Age'] = 2
    dataset.loc[(dataset['Age'] > 22) & (dataset['Age'] <= 27), 'Age'] = 3
    dataset.loc[(dataset['Age'] > 27) & (dataset['Age'] <= 33), 'Age'] = 4
    dataset.loc[(dataset['Age'] > 33) & (dataset['Age'] <= 40), 'Age'] = 5
    dataset.loc[(dataset['Age'] > 40) & (dataset['Age'] <= 66), 'Age'] = 6
    dataset.loc[ dataset['Age'] > 66, 'Age'] = 6
    
data = [train_df, test_df]

for dataset in data:
    dataset.loc[ dataset['Fare'] <= 7.91, 'Fare'] = 0
    dataset.loc[(dataset['Fare'] > 7.91) & (dataset['Fare'] <= 14.454), 'Fare'] = 1
    dataset.loc[(dataset['Fare'] > 14.454) & (dataset['Fare'] <= 31), 'Fare']   = 2
    dataset.loc[(dataset['Fare'] > 31) & (dataset['Fare'] <= 99), 'Fare']   = 3
    dataset.loc[(dataset['Fare'] > 99) & (dataset['Fare'] <= 250), 'Fare']   = 4
    dataset.loc[ dataset['Fare'] > 250, 'Fare'] = 5
    dataset['Fare'] = dataset['Fare'].astype(int)
```

```python
data = [train_df, test_df]
for dataset in data:
    dataset['Age_Class']= dataset['Age']* dataset['Pclass']
```

```python
for dataset in data:
    dataset['Fare_Per_Person'] = dataset['Fare']/(dataset['relatives']+1)
    dataset['Fare_Per_Person'] = dataset['Fare_Per_Person'].astype(int)
```

```python
#The chosen model will be random forest because I'm very familiar with it and I think it applied really well to this situation
#k-fold cross validation will be used in the future to have a more precise idea of how random forest performs
#After that the best random forest parameters will be found
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV, cross_val_score
```

```python
#Data has to be split between target and predictors
X_train = train_df.drop("Survived", axis=1)
y_train = train_df["Survived"]

rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)

#It seems very high thus we're gonna use k-fold cross validation to have a better and more precise idea of the performance
scores = cross_val_score(rf, X_train, y_train, cv=10, scoring = "accuracy")

#Random forest allows us to check the importance of features in the model
importances = pd.DataFrame({'feature':X_train.columns,'importance':np.round(rf.feature_importances_,3)})
importances = importances.sort_values('importance',ascending=False).set_index('feature')
importances.head(9)

#Let's drop 'not_alone' and 'Parch' as it is not informative enough and it can increase the chances of overfitting
train_df  = train_df.drop("not_alone", axis=1)
test_df  = test_df.drop("not_alone", axis=1)

train_df  = train_df.drop("Parch", axis=1)
test_df  = test_df.drop("Parch", axis=1)

X_train = train_df.drop("Survived", axis=1)
y_train = train_df["Survived"]

#Let's train the model again and use this version to make the prediction
rf = RandomForestClassifier(n_estimators=100)
rf.fit(X_train, y_train)

rf = RandomForestClassifier(criterion = "gini", 
                                       min_samples_leaf = 1, 
                                       min_samples_split = 12,   
                                       n_estimators=100, 
                                       max_features='sqrt', 
                                       random_state=1, 
                                       n_jobs=-1)

rf.fit(X_train, y_train)

X_test = test_df.drop("PassengerId", axis = 1)
predictions = rf.predict(X_test)

submission_df = pd.DataFrame({"PassengerId" : test_df["PassengerId"], "Survived" : predictions})
submission_df.to_csv("submission.csv", index=False)
```
